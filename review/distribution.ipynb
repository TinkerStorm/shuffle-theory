```python
import math
from collections import namedtuple, Counter

# from types import 

Role = namedtuple('Role', 'name, ratio')

ROLES = [
  Role("Citizen", 0.5),
  Role("Werewolf", 0.3),
  Role("Spotter", 0.1),
  Role("Hunter", 0.1),
  Role("Seer", 0.1),
  Role("Robber", 0.1),
  Role("Troublemaker", 0.1),
  Role("Drunk", 0.1),
  Role("Insomniac", 0.1)
]

MIN = -10000
MAX = 10000
```

```python
# iterate over players, then gradually increase role count from 2 to max
results = []

for player_count in range(MIN, MAX + 1):
  for role_count in range(2, len(ROLES) + 1):
    # calculate the total ratio of all roles
    ratio_total = sum(role.ratio for role in ROLES[:role_count])
    # calculate the scale factor
    scale_factor = player_count / ratio_total
    # calculate the number of players per role
    roles = []
    acc = 0
    for role_index in range(role_count):
      role = ROLES[role_index]
      count = round(role.ratio * scale_factor)
      acc += count
      for _ in range(count):
        roles.append(role.name)
    # print the number of players that are not accounted for
    results.append({
      'player_count': player_count,
      'role_count': role_count,
      'difference': acc - player_count,
      'difference_percent': abs((acc - player_count) / player_count * 100 if player_count != 0 else 0),
      'difference_log': math.log1p(abs(acc - player_count)),
    })
```

```python
# output the results in a readable format

print("{},{},{},{},{}".format('player_count', 'role_count', 'difference', 'difference_percent', 'difference_log'))
for result in results:
  print("{},{},{},{},{}".format(result['player_count'], result['role_count'], result['difference'], result['difference_percent'], result['difference_log']))
```
